## Code :

## Importing all Necessary Libraries and Loading the Dataset

```python
#import all modules relevant to that
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data[:, :2] 
```

```python
#To ignore warnings
import warnings as wg
wg.filterwarnings("ignore")
```

## To Determine the Optimal number of Clusters.

```python
Elbow Method K-means : It is a method of clustering the dataset for a range of values for k (say from 1-10) and then for each value of k computes an average score of all clusters.
```

```python
wcss = []
from sklearn.cluster import KMeans 

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
From the above, we can observe that WCSS is decreasing as the number of clusters increasing. It represents that WCSS is inversely proportional to number of clusters.
```

## Applying K-Means Method to Visualize Clusters : 

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=pred_y)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.title('K-means Clustering')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

```python
In this way, We can successfully visualise the clustering of data by using k-means algorithm
```
